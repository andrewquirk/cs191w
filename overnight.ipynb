```python
i_citizen = model.distance('immigrant', 'citizen')
e_citizen = model.distance('emigrant', 'citizen')
r_citizen = model.distance('refugee', 'citizen')
m_citizen = model.distance('migrant', 'citizen')

h_citizen = model.distance('human', 'citizen')

N = 5
menMeans = (i_citizen, e_citizen, r_citizen, m_citizen)
hum_citizen = (h_citizen)
ind = np.arange(N-1)
ind5 = np.arange(N)
width = 0.35 
p1 = plt.bar(ind, menMeans, width, label="Migration-Related")
p2 = plt.bar(4, hum_citizen, width, label="Baseline")
plt.ylabel('Similarity')
plt.title('Similarities to \'citizen\'')
plt.xticks(ind5, ('immigrant', 'emigrant', 'refugee', 'migrant', 'person'))

plt.ylim(.5, 1) 

plt.legend(bbox_to_anchor=(.60, .20), loc=2, borderaxespad=0.)
plt.show()


# print(model.distance('refugee', 'person'))
# print(model.distance('migrant', 'person'))
```

```python
comp_vars = [("man", "woman"),("poor", "rich"), ("educated", "uneducated"), ("old", "young"), ("innocent", "criminal"), ("temporary", "permanent"), ("black", "white"), ("african", "european"), ("mexican", "canadian")]

comp_vars2 = [("wanted", "unwanted"), ("safe", "dangerous"), ("crisis", "opportunity"), ("bad", "good"), ("employed", "unemployed")]
for word1, word2 in comp_vars2:
    i_1 = model.distance('immigrant', word1)
    e_1 = model.distance('emigrant', word1)
    r_1 = model.distance('refugee', word1)
    m_1 = model.distance('migrant', word1)
    h_1 = model.distance('human', word1)
    
    i_2 = model.distance('immigrant', word2)
    e_2 = model.distance('emigrant', word2)
    r_2 = model.distance('refugee', word2)
    m_2 = model.distance('migrant', word2)
    h_2 = model.distance('human', word2)
    
    N = 5
    width = 0.35 
    one, two = (i_1, e_1, r_1, m_1), (i_2, e_2, r_2, m_2)
    hum_1, hum_2 = (h_1), (h_2)
    ind = np.arange(N-1)
    ind5 = np.arange(N)
    
    fig, ax = plt.subplots()
    
    
    rects1 = ax.bar(ind - width/2, one, width, 
                    color='SkyBlue', label=word1)
    rects2 = ax.bar(ind + width/2, two, width, 
                    color='IndianRed', label=word2)
    
    rects3 = ax.bar(4 - width/2, hum_1, width, 
                    color='steelblue')
    rects4 = ax.bar(4 + width/2, hum_2, width, 
                    color='brown')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Cosine Distance')
    ax.set_title('Scores by group and gender')
    ax.set_xticks(ind5)
    #ax.xticks(ind5, ('immigrant', 'emigrant', 'refugee', 'migrant', 'human'))
    ax.set_xticklabels(('immigrant', 'emigrant', 'refugee', 'migrant', 'human'))
    ax.legend()

    
   
#     p1 = plt.bar(ind, menMeans, width, label="Migration-Related")
#     p2 = plt.bar(4, hum_citizen, width, label="Baseline")
#     plt.ylabel('Scores')
#     plt.title('Similarity to \'employed\'')
#     plt.xticks(ind5, ('immigrant', 'emigrant', 'refugee', 'migrant', 'human'))

    plt.ylim(.5, 1) 

    # plt.legend((handles=[p1, p2]))
               #, ('Migration-Related', 'Baseline')
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

    plt.show()
```
